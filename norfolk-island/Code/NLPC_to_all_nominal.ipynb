```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)


# Import the csv file
data = pd.read_csv('VDL_Norfolk_Island_Penal_Colony_Cleaned.csv')
```

```python
# Convert pp_sentence_years to nominal with 4 quantiles
data['pp_sentence_years'] = pd.qcut(data['pp_sentence_years'], q=4, duplicates='drop')
print(data['pp_sentence_years'].unique())
print()

# Convert trial_month to nominal with 4 quantiles
data['trial_month'] = pd.qcut(data['trial_month'], q=4, duplicates='drop')
print(data['trial_month'].unique())
print()

# Convert trial_year to nominal with 4 quantiles
data['trial_year'] = pd.qcut(data['trial_year'], q=4, duplicates='drop')
print(data['trial_year'].unique())
print()

# Convert pris_ht_pp to nominal with 4 quantiles
data['pris_ht_pp'] = pd.qcut(data['pris_ht_pp'], q=4, duplicates='drop')
print(data['pris_ht_pp'].unique())
print()

# Convert def_age_pp to nominal with 4 quantiles
data['def_age_pp'] = pd.qcut(data['def_age_pp'], q=4, duplicates='drop')
print(data['def_age_pp'].unique())
print()

# Map def_literacy back to nominal
def_literacy_mapping = {
    0: 'Illiterate',
    1: 'Read Little',
    2: 'Read or Both Little',
    3: 'Read and Write',
    4: 'Superior Literacy',
}
data['def_literacy'] = data['def_literacy'].map(def_literacy_mapping)
print(data['def_literacy'].unique())
print()

# Convert Children to nominal with 4 quantiles
data['children_nr'] = pd.qcut(data['children_nr'], q=4, duplicates='drop')
print(data['children_nr'].unique())
print()

# Convert cash_sav_pp to nominal with 4 quantiles
data['cash_sav_pp'] = pd.qcut(data['cash_sav_pp'], q=4, duplicates='drop')
print(data['cash_sav_pp'].unique())
print()

# Map coloffence_info to nominal
coloffence_info_mapping = {
    True: 'Has Colony Offence',
    False: 'No Colony Offence',
}
data['coloffence_info'] = data['coloffence_info'].map(coloffence_info_mapping)
print(data['coloffence_info'].unique())
print()

# Map offence_ni to nominal
offence_ni_mapping = {
    True: 'Has Offence in NI',
    False: 'No Offence in NI',
}
data['offence_ni'] = data['offence_ni'].map(offence_ni_mapping)
print(data['offence_ni'].unique())
print()

# Map death_in_custody_pp to nominal
death_in_custody_mapping = {
    True: 'Died in Custody',
    False: 'Did Not Die in Custody',
}
data['death_in_custody_pp'] = data['death_in_custody_pp'].map(death_in_custody_mapping)
print(data['death_in_custody_pp'].unique())
print()

# Convert length_of_stay_until_probat to nominal with 4 quantiles
data['length_of_stay_until_probat'] = pd.qcut(data['length_of_stay_until_probat'], q=4, duplicates='drop')
print(data['length_of_stay_until_probat'].unique())
print()

# Convert length_of_stay_until_tl to nominal with 4 quantiles
data['length_of_stay_until_tl'] = pd.qcut(data['length_of_stay_until_tl'], q=4, duplicates='drop')
print(data['length_of_stay_until_tl'].unique())
print()

# Map previous_convictions to nominal
previous_convictions_mapping = {
    True: 'Has Previous Convictions',
    False: 'No Previous Convictions',
}
data['previous_convictions'] = data['previous_convictions'].map(previous_convictions_mapping)
print(data['previous_convictions'].unique())
print()
```
