### This script is used to demonstrate the performance of my implementation of Casper
#### by solving the Two Spirals problem

```python
# import libraries
import numpy as np
import torch
import torch.utils.data
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedKFold
from utilities import set_seed
from casper import CasperNet
from casper import train

# make results determinstic
seed = 4660
if seed != None:
    set_seed(seed)

# Device used
device = 'cpu'
```

```python
def generate_two_spirals(n_points, noise=.5):
    n = np.sqrt(np.random.rand(n_points, 1)) * 780 * (2 * np.pi) / 360
    d1x = -np.cos(n) * n + np.random.rand(n_points, 1) * noise
    d1y = np.sin(n) * n + np.random.rand(n_points, 1) * noise
    return (np.vstack((np.hstack((d1x, d1y)), np.hstack((-d1x, -d1y)))),
            np.hstack((np.zeros(n_points), np.ones(n_points))))

X, labels = generate_two_spirals(1000)

data = pd.DataFrame({'0': labels, '1': X[:, 0], '2': X[:, 1]})
plt.title("The plot of the dataset")
plt.scatter(X[:,0], X[:,1], c=labels)
plt.show()
```

```python
# test the model (modified from lab 2)
def test_two_spirals(model, train_data, test_data):
    model.eval()
    device = next(model.parameters()).device
    
    # test on train set
    train_input = train_data[:, 1:]
    train_target = train_data[:, 0]
    inputs = torch.Tensor(train_input).float().to(device)
    targets = torch.Tensor(train_target).long()
    outputs = model(inputs)
    _, predicted = torch.max(outputs, 1)
    print("Confusion matrix for training:")
    print(confusion_matrix(targets.data, predicted.cpu().long().data))
    
    plt.title("Prediction on training set")
    plt.scatter(train_input[:,0], train_input[:,1], c=predicted.cpu())
    plt.show()

    # test on test set
    test_input = test_data[:, 1:]
    test_target = test_data[:, 0]
    inputs = torch.Tensor(test_input).float().to(device)
    targets = torch.Tensor(test_target).long()
    outputs = model(inputs)
    _, predicted = torch.max(outputs, 1)
    print("Confusion matrix for testing:")
    print(confusion_matrix(targets.data, predicted.cpu().long().data))
    
    # print test accuracy
    total = predicted.size(0)
    correct = predicted.cpu().data.numpy() == targets.data.numpy()
    print('Testing Accuracy: %.2f %%' % (100 * sum(correct)/total))
    
    plt.title("Prediction on test set")
    plt.scatter(test_input[:,0], test_input[:,1], c=predicted.cpu())
    plt.show()
```

```python
# Define hyperparameter
hyperparameters = {'max_hidden_neurons': 16,
                  'P': 5,
                  'D': 0.005,
                  'lrs': [0.2, 0.005, 0.001]}

num_classes, input_size = 2, 2

# randomly split data into training set (80%) and testing set (20%)
train_data, test_data, _, _ = train_test_split(data, data.iloc[:,0], test_size=0.2, random_state=seed)
train_data, test_data = np.array(train_data), np.array(test_data)

# initialise network
casper_net = CasperNet(input_size, num_classes).to(device)

# train the model
train(casper_net, train_data, hyperparameters, 'classification', verbose=True)

# test the model
test_two_spirals(casper_net, train_data, test_data)
```

```python
# validate a single hyperparameter combination
def rkf_validator(data, hyperparameters, n_splits, n_repeats, device='cpu', average_score=True):
    
    data = np.array(data)
    input_dim = data.shape[1] - 1
    all_scores = []
    
    rkf = RepeatedKFold(n_splits=n_splits, n_repeats=n_repeats)
    
    # train the model n_splits * n_repeats times
    for i, (train_index, test_index) in enumerate(rkf.split(X)):
        # obtain training and validation set
        data_train, data_val = data[train_index], data[test_index]
            
        # initialise network
        casper_net = CasperNet(input_dim, num_classes).to(device)
            
        # train the model
        train(casper_net, data_train, hyperparameters, "classification")
            
        # set to evaluation mode
        casper_net.eval()
            
        # compute validation accuracy
        inputs = torch.Tensor(data_val[:, 1:]).float().to(device)
        targets = torch.Tensor(data_val[:, 0]).long()
        outputs = casper_net(inputs).cpu()
        _, predicted = torch.max(outputs, 1)
        total = predicted.size(0)
        correct = predicted.cpu().data.numpy() == targets.data.numpy()
        accuracy = sum(correct) / total * 100
        print(f'Fold: {i + 1}/{(n_splits * n_repeats)}   Accuracy: {accuracy:.4f} %', end = '\r')
        all_scores.append(accuracy)
    
    if average_score:
        return np.mean(all_scores)
    else:
        return all_scores
```

```python
# Train the model 80 times with different test-train split

scores_spiral = rkf_validator(data, hyperparameters, 10, 8, device=device,average_score=False)
print(np.mean(scores_spiral),np.std(scores_spiral))

# print the distribution
plt.title("Accuracy distribution of the model over 80 runs")
plt.hist(scores_spiral, color='lightgreen', ec='black', bins=20)
plt.show()
```
